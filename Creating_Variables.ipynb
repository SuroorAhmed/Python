### Variable Memory

Variables can be assigned new values but the order of operations is important if the new assigned value is to be recognized. 

```python
a = 10
b = 6
```

```python
print(a)
```

```python
c = a + b
print(c)
```

```python
b = 8
```

```python
# Here in the memory, b still = 6

print(c)
```

```python
# Here the new value has been updated. 

c = a + b 
print(c)
```
