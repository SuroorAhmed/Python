Logical Operators

```python
(20 > 15) and (30 > 20) 
```

```python
(20 > 15) or ( 20 > 25)
```

```python
not (20 > 15)
```

```python
not (10 > 15)
```

Membership Operators

in.  Returns true if a sequence with the spicified value is present in the object  
not in   Returns true if a sequence with the spicified valaue is not present in the object

```python
x = "I love Python!"

"Python" in x
```

```python
"SQL" in x
```

```python
"SQL" not in x
```
