```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "Amy", "Jane", "Bert", "Pam", "Colin"],
    "Age": [25, 32, 29, 40, 32, 28, 35, 21],
    "City":["London", "New York", "Paris", "Doha", "Rome", "Milan", "Tokyo", "Sydney"]
}

df = pd.DataFrame(data)
```

```python
df["Age"]
```

```python
df[["Name","Age"]]
```

```python
# Filtering Rows
df[df["Age"] > 30]
```

```python
# Creating a new column by performing math on an existing column
df["AgeIn5Years"] = df["Age"] + 5
```

```python
df[["Name","AgeIn5Years"]]
```

```python
df.groupby("City")["Age"].mean()
```

```python
df.groupby("City")["Age"].agg(["sum", "mean", "max"])
```
