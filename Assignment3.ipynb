```python
# missing percentage
missing_percentage = df.isna().mean()*100
print(missing_percentage)
```

```python
# Make a copy of the original DataFrame
df_mv = df.copy()
print (df.copy)
```

```python
# Impute missing values
df_mv['Ozone'] = df_mv['Ozone'].fillna(df_mv['Ozone'].median())
df_mv['Solar.R'] = df_mv['Solar.R'].fillna(df_mv['Solar.R'].mean())
print (df.copy)
```

The median is preferred over the mean when a dataset has outliers or is skewed.
The mean is sensitive to extreme values and can be pulled up or down by them.
The median is more robust and gives a better central value when data isn’t evenly distributed.

```python
# check that missing values are gone
df_mv.isna().sum()
```
