```python
#Sets - have no order so can not be indexed 

daily_pint_sales_Dec={1,2,3,5}
```

```python
type(daily_pint_sales_Dec)
```

```python
daily_pint_sales_Dec
```

```python
daily_pint_sales_Dec={1,3,2,5,12,6}
```

```python
daily_pint_sales_July={10,8,3,5,6,13}
```

```python
# | shows combined unique values

print(daily_pint_sales_Dec | daily_pint_sales_July)
```

```python
# & shows shared values in both sets

print(daily_pint_sales_Dec & daily_pint_sales_July)
```

```python
print(daily_pint_sales_July - daily_pint_sales_Dec)
```

```python
print(daily_pint_sales_Dec - daily_pint_sales_July)
```

```python
# ^ shows values unique to both sets 

print(daily_pint_sales_Dec ^ daily_pint_sales_July)
```
